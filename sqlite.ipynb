```python
import pandas as pd
import sqlite3 as sql
```

```python
!wget "https://github.com/anvarnarz/praktikum_datasets/raw/main/electricity.db"
```

```python
db = sql.connect("electricity.db")
cursor = db.cursor()
```

```python
cursor.execute("select name from sqlite_master where type = 'table'")
cursor.fetchall()
```

```python
cursor.execute("pragma table_info(elect)")
cursor.fetchall()
```

```python
pd.read_sql("select * from elect", db, index_col='index').head()
```

```python
qy = "select * from elect"
countries = pd.read_sql(qy, db)
countries.size
```

```python
'Uzbekistan' in countries['country'].values
```

```python
uzb = pd.read_sql(qy, db)[pd.read_sql(qy, db)['country'] == 'Uzbekistan']
uzb
```

```python
tw = pd.read_sql("select * from elect where year == 2001 and country != 'World' ORDER BY gas DESC", db, index_col='index')
tw.head(10)
```
