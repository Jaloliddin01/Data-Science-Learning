```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df['num-of-cylinders'].unique()
```

```python
dict = {
    'four' : 4,
    'six' : 6,
    'five' : 5,
    'three' : 3,
    'twelve' : 12,
    'two' : 2,
    'eight' : 8
}

df['cylinders'] = df['num-of-cylinders'].map(dict)
df.head()
```

```python
df['price']
```

```python
kurs = 10820
df['price_uzs'] = df['price'].map(lambda sum: sum * kurs)
df.head()
```

```python
df.replace('four', 4)
```

```python
df.replace(dict)
```
